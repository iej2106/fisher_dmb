```python
# import necessary modules
# uncomment to get plots displayed in notebook
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from classy import Class
from scipy.optimize import fsolve
from scipy.interpolate import interp1d
import math
```

## Parameters I used to generate Cl's

## Load Cl's

```python
cl_tt_est = np.load('Cl_tt.npy')
cl_ee_est = np.load('Cl_ee.npy')
cl_kk_est = np.load('Cl_kk.npy')
cl_pp_est = np.load('Cl_pp.npy')

ell_est = np.arange(cl_tt_est.shape[0])
```

## Planck 2018 paper to compare plots with - https://arxiv.org/pdf/1807.06209.pdf

```python
#################
#
# start plotting
#
#################
# Figure 1 in Planck 2018 paper
plt.xlim([2, 2000])
plt.xlabel(r"$\ell$")
plt.ylabel(r"$\ell (\ell+1) C_l^{TT} / 2 \pi \,\,\, [\mu K^{2}]$")
plt.grid()
#

factor_est = (ell_est*(ell_est+1.)/2./math.pi)
plt.plot(ell_est, factor_est*cl_tt_est,'r-',label=r'$\mathrm{cobaya}$')
#
plt.legend(loc='right',bbox_to_anchor=(1.4, 0.5))
```

```python

plt.xlim([2, 2000])
plt.ylim([-10, 50])
plt.xlabel(r"$\ell$")
plt.ylabel(r"$\ell (\ell+1) C_l^{EE} / 2 \pi \,\,\, [\mu K^{2}]$")
plt.grid()
#
factor_est = (ell_est*(ell_est+1.)/2./math.pi)
plt.plot(ell_est, factor_est*cl_ee_est,'r-',label=r'$\mathrm{cobaya}$')
#
plt.legend(loc='right',bbox_to_anchor=(1.4, 0.5))
```

```python
# Figure 2 (bottom) in Planck 2018 paper
plt.xlim([30, 2000])
plt.ylim([-10, 100])
plt.xlabel(r"$\ell$")
plt.ylabel(r"$C_l^{EE} \,\,\, [10^{-5}\mu K^{2}]$")
plt.grid()
#
factor_est = (ell_est*(ell_est+1.)/2./math.pi)
plt.plot(ell_est, cl_ee_est/1e-5,'r-',label=r'$\mathrm{cobaya}$')
#
plt.legend(loc='right',bbox_to_anchor=(1.4, 0.5))
```

```python
# Figure 3 in Planck 2018 paper
plt.xlim([8,3000])
plt.xlabel(r"$\ell$")
plt.ylabel(r"$[\ell (\ell+1)]^{2} C_l^{\phi\phi} / 2 \pi \,\,\, [\times 10^{7} \mu K^{2}]$")
plt.grid()

factor_pp = (1.e7*((ell_est*(ell_est+1.))**2)/2./math.pi)
plt.semilogx(ell_est,factor_pp*cl_pp_est,'r-',label=r'$\mathrm{cobaya}$')

plt.legend(loc='right',bbox_to_anchor=(1.4, 0.5))
```
