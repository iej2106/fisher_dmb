```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S4_0 = np.loadtxt("fisher_data/S4_n=0/n=0.txt", delimiter = " ")
S4_0 = np.transpose(S4_0)
planck_0 = np.loadtxt("fisher_data/planck_n=0/n=0.txt", delimiter = " ")
planck_0 = np.transpose(planck_0)
S4_0_a = np.loadtxt("MCMC_data/S4_Aizhan_n=0.txt", delimiter = " ")
S4_0_a = np.transpose(S4_0_a)
SO_0 = np.loadtxt("fisher_data/SO_n=0/n=0.txt", delimiter = " ")
SO_0 = np.transpose(SO_0)


S4_2 = np.loadtxt("fisher_data/S4_n=2/n=2.txt", delimiter = " ")
S4_2 = np.transpose(S4_2)
S4_2_a = np.loadtxt("MCMC_data/S4_Aizhan_n=0.txt", delimiter = " ")
S4_2_a = np.transpose(S4_2_a)
planck_2 = np.loadtxt("fisher_data/planck_n=2/n=2.txt", delimiter = " ")
planck_2 = np.transpose(planck_2)

S4_4 = np.loadtxt("fisher_data/S4_n=4/n=4.txt", delimiter = " ")
S4_4 = np.transpose(S4_4)
S4_4_a = np.loadtxt("MCMC_data/S4_Aizhan_n=0.txt", delimiter = " ")
S4_4_a = np.transpose(S4_4_a)

S4_6 = np.loadtxt("fisher_data/S4_n=6/n=6.txt", delimiter = " ")
S4_6 = np.transpose(S4_6)
S4_6_a = np.loadtxt("MCMC_data/S4_Aizhan_n=0.txt", delimiter = " ")
S4_6_a = np.transpose(S4_6_a)

S4_m2 = np.loadtxt("fisher_data/S4_n=-2/n=-2.txt", delimiter = " ")
S4_m2 = np.transpose(S4_m2)

S4_m4 = np.loadtxt("fisher_data/S4_n=-4/n=-4.txt", delimiter = " ")
S4_m4 = np.transpose(S4_m4)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12,18))
plt.title(r"2 sigma upper limit for S4",fontsize=20)

          #### n = 0 ####
axs[0, 0].plot(S4_0[1], S4_0[0], 'b.-',label='S4')
axs[0, 0].plot(S4_0_a[1], S4_0_a[0], 'b.--',label='S4 MCMC')
axs[0, 0].plot(planck_0[1], planck_0[0], 'g.-',label='planck')
axs[0, 0].plot(SO_0[1], SO_0[0], 'r.-',label='SO')
axs[0, 0].set_title('n=0')
axs[0, 0].set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')

          #### n = 2 ####
axs[0, 1].plot(S4_2[1], S4_2[0], 'b.-',label='S4')
axs[0, 1].plot(S4_2_a[1], S4_2_a[0], 'b.--',label='S4 MCMC')
axs[0, 1].set_title('n=2')
axs[0, 1].set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')
          #### n = 4 ####
axs[1, 0].plot(S4_4[1], S4_4[0], 'b.-',label='S4')
axs[1, 0].plot(S4_4_a[1], S4_4_a[0], 'b.--',label='S4 MCMC')
axs[1, 0].set_title('n=4')
axs[1, 0].set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')
          #### n = 6 ####
axs[1, 1].plot(S4_6[1], S4_6[0], 'b.-',label='S4')
axs[1, 1].plot(S4_6_a[1], S4_6_a[0], 'b.--',label='S4 MCMC')
axs[1, 1].set_title('n=6')
axs[1, 1].set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')
          #### n = -2 ####
axs[2, 0].plot(S4_m2[1], S4_m2[0], 'b.-',label='S4')
axs[2, 0].set_title('n=-2')
axs[2, 0].set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')
          #### n = -4 ####
axs[2, 1].plot(S4_m4[1], S4_m4[0], 'b.-',label='S4')
axs[2, 1].set_title('n=-4')
axs[2, 1].set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')


#for ax in axs.flat:
    #ax.set(xlabel='mass (GeV)', ylabel=r'cross section [$cm^2$]')
# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
#    ax.label_outer()
    #ax.set(xlim=(1e-3, 1e3))
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.legend()
    
fig.savefig("fisher_plots/upperlimit_crosssection_all.png")
plt.show()
```
