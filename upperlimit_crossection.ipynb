```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S4_0 = np.loadtxt("fisher_data/S4_n=0/n=0.txt", delimiter = " ")
S4_0 = np.transpose(S4_0)
planck_0 = np.loadtxt("fisher_data/planck_n=0/n=0.txt", delimiter = " ")
planck_0 = np.transpose(planck_0)
S4_0_a = np.loadtxt("MCMC_data/S4_Aizhan_n=0.txt", delimiter = " ")
S4_0_a = np.transpose(S4_0_a)
SO_0 = np.loadtxt("fisher_data/SO_n=0/n=0.txt", delimiter = " ")
SO_0 = np.transpose(SO_0)
cv_limited_0 = np.loadtxt("fisher_data/cv-limited_n=0/n=0.txt", delimiter = " ")
cv_limited_0 = np.transpose(cv_limited_0)


S4_2 = np.loadtxt("fisher_data/S4_n=2/n=2.txt", delimiter = " ")
S4_2 = np.transpose(S4_2)
S4_2_a = np.loadtxt("MCMC_data/S4_Aizhan_n=2.txt", delimiter = " ")
S4_2_a = np.transpose(S4_2_a)
planck_2 = np.loadtxt("fisher_data/planck_n=2/n=2.txt", delimiter = " ")
planck_2 = np.transpose(planck_2)
SO_2 = np.loadtxt("fisher_data/SO_n=2/n=2.txt", delimiter = " ")
SO_2 = np.transpose(SO_2)
cv_limited_2 = np.loadtxt("fisher_data/cv-limited_n=2/n=2.txt", delimiter = " ")
cv_limited_2 = np.transpose(cv_limited_2)

planck_4 = np.loadtxt("fisher_data/planck_n=4/n=4.txt", delimiter = " ")
planck_4 = np.transpose(planck_4)
S4_4 = np.loadtxt("fisher_data/S4_n=4/n=4.txt", delimiter = " ")
S4_4 = np.transpose(S4_4)
S4_4_a = np.loadtxt("MCMC_data/S4_Aizhan_n=4.txt", delimiter = " ")
S4_4_a = np.transpose(S4_4_a)

planck_6 = np.loadtxt("fisher_data/planck_n=6/n=6.txt", delimiter = " ")
planck_6 = np.transpose(planck_6)
S4_6 = np.loadtxt("fisher_data/S4_n=6/n=6.txt", delimiter = " ")
S4_6 = np.transpose(S4_6)
S4_6_a = np.loadtxt("MCMC_data/S4_Aizhan_n=6.txt", delimiter = " ")
S4_6_a = np.transpose(S4_6_a)
SO_6 = np.loadtxt("fisher_data/SO_n=6/n=6.txt", delimiter = " ")
SO_6 = np.transpose(SO_6)

S4_m2 = np.loadtxt("fisher_data/S4_n=-2/n=-2.txt", delimiter = " ")
S4_m2 = np.transpose(S4_m2)
SO_m2 = np.loadtxt("fisher_data/SO_n=-2/n=-2.txt", delimiter = " ")
SO_m2 = np.transpose(SO_m2)
planck_m2 = np.loadtxt("fisher_data/planck_n=-2/n=-2.txt", delimiter = " ")
planck_m2 = np.transpose(planck_m2)

planck_m4 = np.loadtxt("fisher_data/planck_n=-4/n=-4.txt", delimiter = " ")
planck_m4 = np.transpose(planck_m4)
S4_m4 = np.loadtxt("fisher_data/S4_n=-4/n=-4.txt", delimiter = " ")
S4_m4 = np.transpose(S4_m4)
SO_m4 = np.loadtxt("fisher_data/SO_n=-4/n=-4.txt", delimiter = " ")
SO_m4 = np.transpose(SO_m4)


# from Li et al NOTE: slighty different fiducial settings
zack_li_sigma_n0_S4 = 0.6e-26
zack_li_sigma_n0_planck = 16e-26
zack_li_sigma_n0_CVlimited = 0.08e-26
zack_li_m_n0 = 1.0
```
