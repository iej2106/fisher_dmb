```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from classy import Class
import math
```

```python
n_power = 0
dm_mass = 1e-4 #GeV
l=5000
step = 5e-28
#8e-7 * pow(dm_mass,1.3)
#2e-26 * pow(dm_mass,0.15)
#4e-19 * pow(dm_mass,0.70) <- n=2 
#6e-13 * pow(dm_mass,1.25) <- n=4
#3e-34 * pow(dm_mass,0.5)  <- n=-2
```

```python
stepsize = [step*.25, step*0.5, step, step*2, step*4, step*8]
#[step*.25, step*0.5, step, step*2, step*4, step*8]
#[step*0.5, step, step*2]
#[1e-18, 2e-18, 4e-18, 8e-18, 16e-18, 32e-18, 64e-18]
#[step*0.5, step, step*2, step*4, step*8, step*16]
print(stepsize)
print(len(stepsize))
print(step)
```

```python
derivative_cl = []
derivative_norm = []
for x in stepsize:

    params_right = {'output': 'tCl pCl lCl','l_max_scalars': l,'lensing': 'no',
                    'omega_dmb': 0.12,
                    'h': 0.6730804,
                    'omega_b': 0.0224,
                    'A_s': 2.2e-9,
                    'n_s': 0.96,
                    'tau_reio': 0.07,
                    'omega_cdm' :1e-22,
                    'm_dmb': dm_mass,'n_dmb': n_power,'sigma_dmb': x,
                    'N_ur': 2.0328,'N_ncdm': 1,'m_ncdm': 0.06,'T_ncdm':0.71611}
    
    params_center = {'output': 'tCl pCl lCl','l_max_scalars': l,'lensing': 'no',
                    'omega_dmb': 0.12,
                    'h': 0.6730804,
                    'omega_b': 0.0224,
                    'A_s': 2.2e-9,
                    'n_s': 0.96,
                    'tau_reio': 0.07,
                    'omega_cdm' :1e-22,
                    'm_dmb': dm_mass,'n_dmb': n_power,'sigma_dmb': 0.0,
                    'N_ur': 2.0328,'N_ncdm': 1,'m_ncdm': 0.06,'T_ncdm':0.71611}
    
    # Computing Cl_TT for the above parameters in class:
    Mright = Class()
    Mright.set(params_right)
    Mright.compute()
    cl_tot_right = Mright.raw_cl(l)
    Mright.struct_cleanup()
    Mright.empty()
    cl_TT_right = cl_tot_right['tt']
    ell_right = cl_tot_right['ell']
    
    M = Class()
    M.set(params_center)
    M.compute()
    cl_tot = M.raw_cl(l)
    M.struct_cleanup()
    M.empty()
    cl_TT = cl_tot['tt']
    ell = cl_tot['ell']
    
    fiducial_norm = ((cl_TT_right - cl_TT) / (x)) /cl_TT
    derivative_norm.append(fiducial_norm)
    
    derivative = [(cl_TT_right[l] - cl_TT[l]) / (x)]
    derivative_cl.append(derivative)
```
